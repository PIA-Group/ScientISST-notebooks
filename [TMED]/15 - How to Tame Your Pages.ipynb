## <div style="color:#00a0e4"> 1. Background </div>


When programming for the web, there is a separation between content, for-
matting and interation. While HTML is the standard to structure content and CSS is used to define the styling, JavaScript (JS) enables one to add interaction
to a document. It can be used for anything from triggering actions in response to
certain events, to dynamically adjust HTML elements properties and/or styling
rules defined through CSS. In this lab we will experiment with JS to learn about
some of its capabilities.

### <div style="color:#00a0e4">  1. Experimenting with Outputs</div>

The most elementary operations with JS are related with presenting output
content to a user or as development data. The goal of this experiment is for you
to get familiar with JS and some of its possibilities:

    1. Create a new (empty) HTML file in the Spyder IDE (or another HTML editor of your choice)

    2. Write a document that has one textbox for data input by the user, four buttons and one paragraph

    3. Program each of the buttons to, when clicked, show the text written by the user respectivelly in the JS console, as an alert message, as the content of the paragaph, and as an output to the HTML document (one operation per button)

    4. Open the HTML file using Google Chrome (if its open already, you can simply refresh the page) to confirm the result

### <div style="color:#00a0e4">  2. Events Aplenty </div>

With JS you can detect a large number of events and perform actions based on
them. With this exercise, the goal is for you to experiment with JS events:

    1. Create a new (empty) HTML file in the Spyder IDE (or another HTML editor of your choice)

    2. Write a document in which the background color switches from black to white whenever the user presses a key

    3. Open the HTML file using Google Chrome (if its open already, you can simply refresh the page) to confirm the result

    4. Expand the document in such way that the transparency of the color changes with the mouse movement

    5. Open the HTML file using Google Chrome (if its open already, you can simply refresh the page) to confirm the result
