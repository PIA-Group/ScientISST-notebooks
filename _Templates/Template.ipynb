### <div style="color:#48ba57">   3.1. Plot Example  </div>

The following color palette should be respected in case any plot or graphic is drawn:

````python
color_palette=['#48ba57', '#fbb144', '#ef564b', '#726658', '#946db2', '#46bd96', '#f26451']
````

````python
# to enable basic visualization tools troughout the whole notebook:
!pip install mpld3 >/dev/null 2>&1
import mpld3 
 
# note: in case you are using subplots, this tool will not work. In those cells, please add the following command:
# mpld3.enable_notebook()

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s1 = 1 + np.sin(2 * np.pi * t)
s2 = 2 + np.sin(1.5 * np.pi * t)
s3 = 2 + np.sin(2 * np.pi * t)

# plot
plt.title('Title of the plot', {'size':20})
plt.xlabel('Time (s)')
plt.ylabel('Angle [ยบ]')
plt.plot(t, s1, label='signal 1',  color=color_palette[0])
plt.plot(t, s2, label='signal 2', color=color_palette[5])
plt.plot(t, s3, label='signal 3', color=color_palette[1])
plt.legend()
plt.show()
````
